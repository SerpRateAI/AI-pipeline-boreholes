```python
df_elliot = pd.read_excel('Datasets/Dataset_BA1B_Final.xlsx')
df_elliot.head()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(10, 15), sharey=True)

df[['Top depth (m downhole)', 'Alteration %']].plot(x='Alteration %', y='Top depth (m downhole)', ax=ax[0])
df[['Top depth (m downhole)', 'Alteration %']].plot(x='Alteration %', y='Top depth (m downhole)', ax=ax[2])
df[['Top depth (m downhole)', 'Alteration %']].plot(x='Alteration %', y='Top depth (m downhole)', ax=ax[3])

ax2 = ax[0].twiny()
df_elliot.plot(x='% of fractures', y='TOP_DEPTH', color='red', ax=ax2)

df[['Top depth (m downhole)', 'Alteration %']].fillna(0).sort_values('Top depth (m downhole)').rolling(5).mean().plot(x='Alteration %', y='Top depth (m downhole)', ax=ax[1])

ax3 = ax[1].twiny()
df_elliot[['% of fractures', 'TOP_DEPTH']].rolling(5).mean().plot(x='% of fractures', y='TOP_DEPTH', color='red', ax=ax3)

ax4 = ax[2].twiny()
df_frac.plot(x='Fracture density', y='Top depth (m downhole)', color='cyan', ax=ax4)

ax5 = ax[3].twiny()
df_veins.plot(x='Vein density', y='Top depth (m downhole)', color='green', ax=ax5)


ax[0].set_ylim(400, 0)
```
